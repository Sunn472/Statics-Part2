```python
'''
Q3. A Write a Python function to calculate the probability distribution function
(PDF) of a Bernoulli distribution
'''
import numpy as np

def bernoulli_pdf(x, p):
    x = np.asarray(x)

    # Validate inputs
    if np.any((x != 0) & (x != 1)):
        raise ValueError("All x values must be 0 or 1")
    if not 0 <= p <= 1:
        raise ValueError("p must be between 0 and 1")

    # Bernoulli PDF: p if x=1, (1-p) if x=0
    return np.where(x == 1, p, 1 - p)


print(bernoulli_pdf(1, 0.7))
```

```python
'''
Q6. A Write a Python program to calculate and plot the cumulative distribution
function (CDF) of a discrete uniform distribution
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

def plot_uniform_cdf(low, high):

    # Create the distribution
    dist = randint(low, high + 1)

    # Generate all possible values
    x_values = np.arange(low - 1, high + 2)
    cdf_values = dist.cdf(x_values)

    # Plotting
    plt.figure(figsize=(10, 6))

    # Step plot for discrete CDF
    plt.step(x_values, cdf_values, where='post', color='blue', linewidth=2.5)

    # Mark the points
    plt.scatter(x_values[1:], cdf_values[1:], color='red', s=50, zorder=3)

    # Customize the plot
    plt.title(f'CDF of Discrete Uniform Distribution ({low} to {high})')
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.xticks(np.arange(low, high + 1))
    plt.yticks(np.linspace(0, 1, 11))
    plt.grid(True, alpha=0.3)
    plt.ylim(0, 1.05)

    # Annotate the jumps
    for x, y in zip(x_values[1:-1], cdf_values[1:-1]):
        plt.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=10)

    plt.show()

# Example usage
plot_uniform_cdf(low=1, high=6)
```

```python
'''
Q15. Implement hypothesis testing using Z-statistics for a sample dataset
'''

def z_test(sample, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    sample_mean = np.mean(sample)
    n = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std/np.sqrt(n))

    if alternative == 'two-sided':
        p_value = 2 * norm.sf(np.abs(z_score))
    elif alternative == 'greater':
        p_value = norm.sf(z_score)
    else:  # 'less'
        p_value = norm.cdf(z_score)

    reject = p_value < alpha
    return z_score, p_value, reject

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=105, scale=15, size=30)  # Sample with true mean=105
z, p, reject = z_test(sample_data, pop_mean=100, pop_std=15, alpha=0.05, alternative='two-sided')

print(f"Z-score: {z:.3f}, p-value: {p:.4f}")
print(f"Reject null hypothesis at =0.05: {reject}")
```

```python
'''
Q16. Create a confidence interval for a dataset using Python and interpret
the result
'''
import numpy as np
from scipy.stats import norm, t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Sample standard deviation

    if n >= 30:  # Z-interval for large samples
        margin = norm.ppf((1 + confidence)/2) * std/np.sqrt(n)
    else:  # t-interval for small samples
        margin = t.ppf((1 + confidence)/2, df=n-1) * std/np.sqrt(n)

    return (mean - margin, mean + margin)

# Example usage
data = np.random.normal(loc=50, scale=10, size=100)  # Sample dataset
ci = confidence_interval(data, confidence=0.95)

print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
print(f"Interpretation: We're 95% confident the true population mean falls between {ci[0]:.2f} and {ci[1]:.2f}")
```

```python
'''
Q24. Write a Python function to calculate the skewness and kurtosis of a
dataset generated from a normal distribution.
'''
from scipy.stats import skew, kurtosis

def normal_stats(data):
    """Calculate skewness and kurtosis of a normally distributed dataset."""
    return {
        'skewness': skew(data),
        'kurtosis': kurtosis(data, fisher=True)  # Fisher's definition (normal=0)
    }

# Example usage
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)
stats = normal_stats(data)

print(f"Skewness: {stats['skewness']:.3f} (0 for symmetry)")
print(f"Kurtosis: {stats['kurtosis']:.3f} (0 for normal tails)")
```

```python
'''
Q1. Write a Python program to perform a Z-test for comparing a sample mean to a
known population mean and interpret the results
'''
def z_test(sample, pop_mean, pop_std, alpha=0.05):
    sample_mean = np.mean(sample)
    n = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * norm.sf(np.abs(z_score))  # Two-tailed test
    reject = p_value < alpha
    return z_score, p_value, reject

# Example usage
np.random.seed(42)
pop_mean = 100      # Known population mean
pop_std = 15        # Known population std
sample = np.random.normal(loc=105, scale=pop_std, size=30)  # Sample with true mean=105

z, p, reject = z_test(sample, pop_mean, pop_std)

print(f"Z-score: {z:.3f}\np-value: {p:.4f}")
print(f"Reject H0 (={pop_mean})? {'Yes' if reject else 'No'}")
```

```python
'''
Q3. Implement a one-sample Z-test using Python to compare the sample mean with
the population mean
'''

def z_test(sample, pop_mean, pop_std, alpha=0.05, two_tailed=True):
    sample_mean = np.mean(sample)
    n = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    if two_tailed:
        p_value = 2 * norm.sf(np.abs(z_score))
    else:
        p_value = norm.sf(z_score) if z_score > 0 else norm.cdf(z_score)

    reject = p_value < alpha
    return z_score, p_value, reject

np.random.seed(42)
pop_mean = 100
pop_std = 15
sample = np.random.normal(loc=105, scale=pop_std, size=30)

alpha = 0.05  # Define alpha before using it in the print statement
z, p, reject = z_test(sample, pop_mean, pop_std, alpha=alpha)

print(f"Z-score: {z:.3f}\np-value: {p:.4f}")
print(f"Reject H0? {'Yes' if reject else 'No'} (={alpha})")
```

```python
'''
Q5. Create a Python function that calculates and visualizes Type 1 and
Type 2 errors during hypothesis testing
'''

def visualize_errors(true_mean=100, effect_size=5, std=15, n=30, alpha=0.05):
    # Generate distributions
    null_dist = norm(loc=true_mean, scale=std/np.sqrt(n))
    alt_dist = norm(loc=true_mean + effect_size, scale=std/np.sqrt(n))
    critical_val = null_dist.ppf(1 - alpha)

    # Calculate errors
    type1 = alpha  # P(Reject H0 | H0 true)
    type2 = alt_dist.cdf(critical_val)  # P(Fail to reject H0 | H1 true)
    power = 1 - type2

    # Plot
    x = np.linspace(true_mean-4*std/np.sqrt(n), true_mean+6*std/np.sqrt(n), 1000)
    plt.figure(figsize=(10,5))

    plt.plot(x, null_dist.pdf(x), label='Null (H0)')
    plt.plot(x, alt_dist.pdf(x), label=f'Alternative (H1, ={effect_size})')

    # Type 1 error ()
    plt.fill_between(x, null_dist.pdf(x), where=(x > critical_val),
                    color='red', alpha=0.3, label=f'Type 1 (={type1:.3f})')

    # Type 2 error ()
    plt.fill_between(x, alt_dist.pdf(x), where=(x < critical_val),
                    color='blue', alpha=0.3, label=f'Type 2 (={type2:.3f})')

    plt.axvline(critical_val, color='k', linestyle='--', label=f'Critical value')
    plt.title(f'Type 1 vs Type 2 Errors\nPower = {power:.3f}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(alpha=0.2)
    plt.show()

# Example usage
visualize_errors(true_mean=100, effect_size=5, std=15)
```

```python
'''
Q9. Write a Python function to calculate the confidence interval for a
sample mean and explain its significance
'''
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # Standard error

    # Use t-distribution for small samples, normal for large
    if n < 30:
        crit_val = t.ppf((1 + confidence)/2, df=n-1)  # t-score
    else:
        crit_val = norm.ppf((1 + confidence)/2)       # z-score

    margin = crit_val * std_err
    return (mean - margin, mean + margin), margin

# Example usage
sample = np.random.normal(loc=50, scale=10, size=20)
ci, margin = confidence_interval(sample)
print(f"95% CI: ({ci[0]:.2f}, {ci[1]:.2f})")
print(f"Margin of error: {margin:.2f}")
print(f"Interpretation: We're 95% confident the true population mean is between {ci[0]:.2f} and {ci[1]:.2f}")
```

```python
'''
Q10. Write a Python program to calculate the margin of error for a given
confidence level using sample data
'''
import numpy as np
from scipy.stats import norm, t

def margin_of_error(data, confidence=0.95):
    n = len(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    std_err = sample_std / np.sqrt(n)  # Standard error

    # Use t-distribution for small samples, normal for large
    if n < 30:
        crit_val = t.ppf((1 + confidence)/2, df=n-1)  # t-critical value
    else:
        crit_val = norm.ppf((1 + confidence)/2)       # z-critical value

    return crit_val * std_err

# Example usage
sample_data = np.random.normal(loc=50, scale=10, size=25)  # Sample data
mo_error = margin_of_error(sample_data, confidence=0.95)

print(f"Margin of error (95% CI): {mo_error:.2f}")
print(f"Interpretation: The true mean is within {mo_error:.2f} of the sample mean with 95% confidence")
```

```python
'''
Q19. Write a Python function to check the assumptions (normality, independence,
and equal variance) for ANOVA
'''

import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(groups, alpha=0.05):
    results = {}

    # 1. Normality check (Shapiro-Wilk for each group)
    print("=== Normality Check ===")
    for i, group in enumerate(groups, 1):
        stat, p = stats.shapiro(group)
        is_normal = p > alpha
        results[f'Group_{i}_normality'] = {'statistic': stat, 'p-value': p, 'is_normal': is_normal}
        print(f"Group {i}: {'Normal' if is_normal else 'Non-normal'} (p={p:.3f})")

    # 2. Equal variance check (Levene's test)
    print("\n=== Equal Variance Check ===")
    levene_stat, levene_p = stats.levene(*groups)
    equal_var = levene_p > alpha
    results['homogeneity_var'] = {'statistic': levene_stat, 'p-value': levene_p, 'equal_variances': equal_var}
    print(f"Equal variances: {'Yes' if equal_var else 'No'} (p={levene_p:.3f})")

    # 3. Independence (no test, just reminder)
    print("\n=== Independence Check ===")
    print("Independence is assumed based on experimental design")

    # QQ plots
    plt.figure(figsize=(12, 4))
    for i, group in enumerate(groups, 1):
        plt.subplot(1, len(groups), i)
        stats.probplot(group, plot=plt)
        plt.title(f'QQ Plot - Group {i}')
    plt.tight_layout()
    plt.show()

    return results

group1 = [25, 30, 28, 36, 29]
group2 = [45, 55, 29, 56, 40]
group3 = [30, 29, 33, 37, 27]
check_anova_assumptions([group1, group2, group3])
```

```python
'''
Q24. Perform a hypothesis test for population variance using a Chi-square
distribution and interpret the results
'''

# Sample data - normally distributed with =25 (=5)
np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)
sample_var = np.var(sample, ddof=1)  # sample variance (unbiased)

# Hypothesis test parameters
H0_var = 25  # null hypothesis population variance
alpha = 0.05  # significance level

# Calculate Chi-square statistic
chi2_stat = (len(sample) - 1) * sample_var / H0_var

# Two-tailed test critical values and p-value
lower_crit = chi2.ppf(alpha/2, df=len(sample)-1)
upper_crit = chi2.ppf(1-alpha/2, df=len(sample)-1)
p_value = 2 * min(chi2.cdf(chi2_stat, df=len(sample)-1),
                  1 - chi2.cdf(chi2_stat, df=len(sample)-1))

# Results
print(f"Sample variance: {sample_var:.2f}")
print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"Critical region: < {lower_crit:.2f} or > {upper_crit:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if chi2_stat < lower_crit or chi2_stat > upper_crit:
    print(f"Reject H (p = {p_value:.4f}): Population variance  {H0_var}")
else:
    print(f"Fail to reject H (p = {p_value:.4f}): No evidence variance differs from {H0_var}")
```
